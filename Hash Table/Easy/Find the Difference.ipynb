# <span style='font-family:"Times New Roman"'> <span styel=''> Find the Difference

## <span style='font-family:"Times New Roman"'> <span styel=''> You are given two strings s and t.

## <span style='font-family:"Times New Roman"'> <span styel=''> String t is generated by random shuffling string s and then add one more letter at a random position.

## <span style='font-family:"Times New Roman"'> <span styel=''> Return the letter that was added to t.

```python
import collections
```

```python
def findTheDifference(s, t):
    count_s = collections.Counter(s)
    count_t = collections.Counter(t)
    for key,value in count_t.items():
        if key not in count_s.keys() or count_t[key] != count_s[key]:
            return(key)
```

```python
s = "abcd"
t = "abcde"
print('Output: ' + str(findTheDifference(s, t)))
```

```python
s = ""
t = "y"
print('Output: ' + str(findTheDifference(s, t)))
```

```python
s = "a"
t = "aa"
print('Output: ' + str(findTheDifference(s, t)))
```
